#GoPlay Profile

##Background

GoPlay is Indonesia’s live streaming service with 1000+ creator partners hosting online live sessions to their viewers and fans. Creator partners are spread across genres (e.g. beauty, health, gaming, religion, and many others) and their live sessions are recorded, enabling viewers to replay them at their convenience. Creators earn money through optional online tips given by their fans and viewers in the form of GoPlay Tokens. GoPlay audiences can buy Tokens. Each Token costs one Indonesian Rupiah.

GoPlay identifies hero creators as creator partners that contribute significantly to its live streaming services, which is purely determined by the monetary contribution they provide to the company. If their monthly tips or revenue surpass the monthly minimum wage of Jakarta, content creators become hero creators. . 

A Hero Creator is a creator who makes more than IDR 4.5million in revenue.

##Data Description:

Viewer Percentile: The percentage comparison of the creator’s total views to their peers’. A score percentile of X means that the creator’s total views on their videos exceeded x%  of all creators. For example, a creator on the 95% viewer percentile means that he/she is top 5% among all creators in terms of total number of viewers on his/her videos.

Creator’s Average Rating: Average score received by the content creator, ranging from 1 to 5.

Verified: A verified creator receives a special “checkmark” symbol on his/her account profile, indicating that GoPlay has formally recognized the significance/value of the creator’s  contribution. 

Gender: The creator’s gender, which is either male or female

Average Session per Week (last month): The average livestreaming videos performed per week in a month.

Genre: The main theme of a creator’s content (only 4 possible values): Hobbies/Interests, Sports/Gaming, Religion, or Entertainment.

Revenue: The creator’s gross income in February 2022 from live streaming in GoPlay’s platform.

```python
import matplotlib.pyplot as plt
df.groupby(['Gender','Verified','Genre'])['Revenue'].mean().unstack().plot(kind='bar')
plt.title('Average of Revenue', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Genre', fontsize = 12)
plt.ylabel('Revenue (in Billions)', fontsize = 12)
plt.legend(bbox_to_anchor=(1,1), shadow=True, title='Genre')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The result of the bar chart of revenue by genre and gender is the most Entertainment/Talkshows genre with gender Male in each genre and verified accounts
